# Normalverteilungen erzeugen und plotten

```python
import numpy as np
x = np.random.normal(15, 1, 1000)
y = np.random.normal(15, 1.5, 1000)

x = np.append(x, np.random.normal(10, 1, 1000))
y = np.append(y, np.random.normal(10, 1, 1000))

x = np.append(x, np.random.normal(15, 0.5, 1000))
y = np.append(y, np.random.normal(8, 1, 1000))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = set(["retina"])
plt.scatter(x, y, s=0.2)
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
