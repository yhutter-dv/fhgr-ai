# Lorenz-Funktion und Gini-Koeffizient

Beispiel einer Einkommensverteilung einer Bevölkerung

```python
import numpy as np
person = np.arange(1,100,1)
einkommen = [0]
#einkommen = person + 10000 # fast gleichverteiltes Einkommen
einkommen = 1500 + person  * 1000.2 + einkommen #Ziemlich ungleiche Einkommensverteilung
#einkommen = 1500 + (person  * person * person * person) + einkommen #extrem ungleiche Einkommensverteilung

einkommen
```

Sortierte Einkommensverteilung

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = set(["retina"])
plt.plot(person, einkommen, linewidth = 3)
plt.xlabel("Person")
plt.ylabel("Einkommen [€]")
plt.show()
```

```python
einkommensmenge = sum (einkommen)
einkommensmenge
```

Lorenz-Funktion

```python
einkommen_summiert = 0
x = np.array([])
y = np.array([])
for i in np.arange(1,100,1):
    einkommen_summiert += einkommen[i-1]
    einkommensanteil = (einkommen_summiert / einkommensmenge) * 100 #%
    x = np.append(x, i)
    y = np.append(y, einkommensanteil)
    
```

```python

plt.plot(x, y, linewidth = 3) #Lorenz-Funktion
plt.plot(x, x, linewidth = 1, color='black') # Ideal-Referenz

plt.xlabel("Bevölkerungsanteil [%]")
plt.ylabel("Einkommensmenge [%]")
plt.show()
```

Gini-Fläche

```python
plt.plot(x, y, linewidth = 1)
plt.plot(x, x, linewidth = 3, color='black')
plt.xlabel("Bevölkerungsanteil [%]")
plt.ylabel("Einkommensmenge [%]")
plt.fill_between(x, y, color='red') #Gini-Fläche (rot)
plt.show()
```

### Gini-Koeffizient
Fläche unter rot im Verhältnis zur Fläche unter schwarz

```python
Fläche_unter_der_Kurve = sum(y)
Ideale_Gini_Fläche = 100 * 100 / 2   # 100% * 100* / 2 -> Fläche unter schwarz
Gini_Koeffizient = 1 - (Fläche_unter_der_Kurve / Ideale_Gini_Fläche)
Gini_Koeffizient
```
